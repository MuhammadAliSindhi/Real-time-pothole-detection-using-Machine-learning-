```python
import cv2
import numpy as np
from PIL import Image
from keras import models
from datetime import datetime
```

```python
model=models.load_model("model_saved/model.h5")   
```

```python

video = cv2.VideoCapture(0)
while True:
        now = datetime.now()
        dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
        _, frame = video.read()

        #Convert the captured frame into RGB
        im = Image.fromarray(frame, 'RGB')

        #Resizing into 128x128 because we trained the model with this image size.
        im = im.resize((128,128))
        img_array = np.array(im)
        #print(img_array.shape)

        #Our keras model used a 4D tensor, (images x height x width x channel)
        #So changing dimension 128x128x3 into 1x128x128x3 
        img_array = np.expand_dims(img_array, axis=0)
       # print(img_array)
        #Calling the predict method on model to predict 'pothole' on the image
        prediction = int(model.predict(img_array)[0][0])
        #if prediction is 0, which meanspothole is missing on the image, then show the frame in gray color.
        if prediction == 0:
            print("There is no potholes in this area.")
            cv2.imshow("Detecteding potholes", frame)
        else:
            cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            
            print("There is potholes in this area:")
            print("pothole detected at: ",dt_string)
            
        key=cv2.waitKey(1)
        if key == ord('q'):
                break
video.release()
cv2.destroyAllWindows()    
```

```python
import requests


def get_ip():
    response = requests.get('https://api64.ipify.org?format=json').json()
    return response["ip"]


def get_location():
    ip_address = get_ip()
    response = requests.get(f'https://ipapi.co/{ip_address}/json/').json()
    location_data = {
        "ip": ip_address,
        "city": response.get("city"),
        "region": response.get("region"),
        "country": response.get("country_name")
    }
    return location_data


print(get_location())
```
